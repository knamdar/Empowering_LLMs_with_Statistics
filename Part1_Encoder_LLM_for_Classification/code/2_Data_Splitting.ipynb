## Data Splitting
Create a 10-fold stratified random data splitting based on Label and ID, saved as `Folds.csv`.

````python
import pandas as pd
from sklearn.model_selection import StratifiedKFold
import os

data_path = "../data/AIS.csv"
output_path = "../data/Folds.csv"

df = pd.read_csv(data_path)

skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
folds = []

for fold, (train_idx, test_idx) in enumerate(skf.split(df['ID'], df['Label'])):
    test_ids = df.loc[test_idx, 'ID'].values
    for test_id in test_ids:
        folds.append({'ID': test_id, 'Fold': fold})

folds_df = pd.DataFrame(folds)
folds_df.sort_values(by='ID', inplace=True)
folds_df.to_csv(output_path, index=False)
print(f"Folds saved to {output_path}")
````
